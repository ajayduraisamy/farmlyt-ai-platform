```python
# =========================
# 1. Mount Google Drive
# =========================
from google.colab import drive
drive.mount('/content/drive')

# =========================
# 2. Install Ultralytics
# =========================
!pip install -q ultralytics

# =========================
# 3. Import YOLO
# =========================
from ultralytics import YOLO

# =========================
# 4. Load Model
# =========================
# Option A (Recommended): Strong base model
model = YOLO("yolov8m.pt")

# Option B (your custom pretrained)
# model = YOLO("/content/drive/MyDrive/brinjal_100_dataset/brinjal_leaf_model_v1.pt")

# =========================
# 5. Train Model
# =========================
model.train(
    data="/content/drive/MyDrive/brinjal_100_dataset/data.yaml",

    # Training params
    epochs=150,              # increased for better learning
    imgsz=640,
    batch=16,                # increase if GPU allows

    # Optimization
    optimizer="AdamW",
    lr0=0.001,               # better than 0.0003 for your dataset
    weight_decay=0.0005,

    # Early stopping
    patience=30,             # allow more learning

    # Augmentation
    augment=True,
    degrees=5.0,
    translate=0.1,
    scale=0.5,
    shear=2.0,
    fliplr=0.5,

    # Performance
    workers=8,
    cache=True,              # faster training

    # Logging
    project="/content/drive/MyDrive/brinjal_100_results",
    name="brinjal_final_v3",
    exist_ok=True,

    # Validation
    val=True,
    plots=True
)

# =========================
# 6. Validate Best Model
# =========================
model.val()

# =========================
# 7. Predict (Test)
# =========================
model.predict(
    source="/content/drive/MyDrive/test_images",  # your test folder
    conf=0.25,
    save=True
)
```
